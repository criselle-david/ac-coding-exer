```python
from glob import glob
import pandas as pd
import re
import os
```

```python
files = glob('/mnt/data/public/census/_*')
```

```python
output_folder = '../output'
```

```python
labels = list(map(lambda x: x.split('_')[1], files))
labels
```

```python
label_map = {}
for label in labels:
    if label == 'CARAGA':
        label_map[label] = 'REGION XIII'
    elif label == 'MIMAROPA':
        label_map[label] = 'REGION IV-B'
    else: 
        label_map[label] = label
label_map
```

```python
from collections import OrderedDict
sheet_names = pd.ExcelFile(glob('/mnt/data/public/census/_*')[0]).sheet_names
table_names = [pd.read_excel(glob('/mnt/data/public/census/_*')[0], i, header=None).iloc[0,0] for i in sheet_names]
tables = OrderedDict(zip(sheet_names, table_names))
```

```python
table_names
```

# religion per region

```python
df_religion = pd.DataFrame()
for file, region in zip(files, labels):
    df = pd.read_excel(file, sheet_name='T8', header=2).dropna(how='all').iloc[2:-1, :]
    df = df[['Religious Affiliation', 'Both Sexes']].copy()
    df['region'] = len(df) * [region]
    df_religion = pd.concat([df_religion, df], axis=0)
#     df = df.set_index('Religious Affiliation')
    # df = df.dropna(how='all')
#     df_religion[label_map[region]] = df['Both Sexes']
# df_religion = df_religion.T.copy()
df_religion.columns = ['affiliation', 'population', 'region']
df_religion.to_csv(os.path.join(output_folder, f'census_religion.csv'),index=False)
```
